# Figures for manuscript: spliceAI boxplot and consensus splice sites sequences

## SpliceAI Boxplot

````python
import seaborn as sns

file_in = "/data2/Exome_analysis_BC/Server_S_exome/RNA-Seq_bioinformatique/RNU4-2/rmats/PUROMOINS/results_25072024_19RNU/A5SS.19RNU.df.diff_delta_v1.txt"
df_box = pd.DataFrame(columns=["severity","type","spliceai"])
a = 0
with open(file_in,'r') as f:
    for line in f:
        a+=1
        if a == 1:
            continue
        field = line.split('\t')
        gene = field[0]
        severity = field[7]
        spliceai_cano = field[8]
        if spliceai_cano == 'NA':
            continue
        spliceai_cano = float(spliceai_cano)
        spliceai_alt = float(field[9])
        df_box.loc[gene+"_inc"] = [severity,'Increased',spliceai_alt]
        df_box.loc[gene+"_dec"] = [severity,'Decreased',spliceai_cano]

plt.figure(figsize=(6,6))
ax = sns.boxplot(data=df_box,palette="pastel", hue="severity", x="type", y="spliceai", order=["Decreased", "Increased"])
ax.set_xticklabels(['Decreased 5SS','Increased 5SS'])

#ax.set_xticklabels(['Decreased site','Increased site','de','de'])
#ax.set(ylabel='Raw spliceAI scores')
plt.savefig("boxplot_rnu42.svg", format='svg',transparent=True)
plt.show()
````

## T-test for boxplot

````python
from scipy.stats import ttest_ind
df_box_decreased_severe = df_box[(df_box["type"]=="Decreased") & (df_box["severity"]=="TRUE")]["spliceai"]
df_box_decreased_shared = df_box[(df_box["type"]=="Decreased") & (df_box["severity"]=="FALSE")]["spliceai"]
print("t-test for decreased sites: ", ttest_ind(df_box_decreased_severe, df_box_decreased_shared))

df_box_increased_severe = df_box[(df_box["type"]=="Increased") & (df_box["severity"]=="TRUE")]["spliceai"]
df_box_increased_shared = df_box[(df_box["type"]=="Increased") & (df_box["severity"]=="FALSE")]["spliceai"]
print("t-test for increased sites: ", ttest_ind(df_box_increased_severe, df_box_increased_shared))
````

````python
print("Median decreased severe: "+ str(np.median(df_box[(df_box["type"]=="Decreased") & (df_box["severity"]=="TRUE")]["spliceai"])))
print("Median decreased shared: "+ str(np.median(df_box[(df_box["type"]=="Decreased") & (df_box["severity"]=="FALSE")]["spliceai"])))
print("Median increased severe: "+ str(np.median(df_box[(df_box["type"]=="Increased") & (df_box["severity"]=="TRUE")]["spliceai"])))
print("Median increased shared: "+ str(np.median(df_box[(df_box["type"]=="Increased") & (df_box["severity"]=="FALSE")]["spliceai"])))
````
